```python
import random
```

```python
numbers = list(range(1, 10001))
random.shuffle(numbers)
print(numbers)
```

```python
actual_label = {"Class 1":[],"Class 2":[],"Class 3":[],"Class 4":[],"Class 5":[]}

# Populate the dictionary
for i in range(1, 10001):
    if i <= 2000:
        actual_label["Class 1"].append(i)
    elif i <= 4000:
        actual_label["Class 2"].append(i)
    elif i <= 6000:
        actual_label["Class 3"].append(i)
    elif i <= 8000:
        actual_label["Class 4"].append(i)
    else:
        actual_label["Class 5"].append(i)

# Print the dictionary
for key, value in actual_label.items():
    print(key, value[:5], "...") # Print only the first five numbers of each class for brevity
```

```python
predicted_label = {"Class 1":[],"Class 2":[],"Class 3":[],"Class 4":[],"Class 5":[]}

# Populate the dictionary
for i in range(0, 10000):
    if i <= 2000:
        predicted_label["Class 1"].append(random.randint(1, 10000))
    elif i <= 4000:
        predicted_label["Class 2"].append(random.randint(1, 10000))
    elif i <= 6000:
        predicted_label["Class 3"].append(random.randint(1, 10000))
    elif i <= 8000:
        predicted_label["Class 4"].append(random.randint(1, 10000))
    else:
        predicted_label["Class 5"].append(random.randint(1, 10000))
# Print the dictionary
for key, value in predicted_label.items():
    print(key, value[:5], "...") # Print only the first five numbers of each class for brevity
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
```

```python
predicted_dict =  predicted_label
class_ranges = {
    "Class 1": range(1,2001),
    "Class 2": range(2001,4001),
    "Class 3": range(4001,6001),
    "Class 4": range(6001,8001),
    "Class 5": range(8001,10001)
}
```

```python
# Initialize confusion matrix with zeros
confusion_matrix = np.zeros((len(predicted_dict), len(predicted_dict)))

# Fill the confusion matrix
for i, true_class in enumerate(class_ranges):
    for j, pred_class in enumerate(predicted_dict):
        confusion_matrix[i,j] = len(set(class_ranges[true_class]).intersection(set(predicted_dict[pred_class])))
```

```python
# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix,
                              display_labels=predicted_dict.keys())

disp.plot(include_values=True,
          cmap='viridis', ax=None,
          xticks_rotation='horizontal',
          values_format='.0f')
plt.show()
```
